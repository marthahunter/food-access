```python
census_df['pct_under_10k'] = round((census_df['less_than_10k']/census_df['total']*100),2)
census_df['pct_10k_to_14.9k'] = round((census_df['10k_to_14.999k']/census_df['total']*100),2)
census_df['pct_15k_to_19.9k'] = round((census_df['15k_to_19.999k']/census_df['total']*100),2)
census_df['pct_20k_to_24.9k'] = round((census_df['20k_to_24.999k']/census_df['total']*100),2)
census_df['pct_25k_to_29.9k'] = round((census_df['25k_to_29.999k']/census_df['total']*100),2)
census_df['pct_30k_to_34.9k'] = round((census_df['30k_to_34.999k']/census_df['total']*100),2)
census_df['pct_35k_to_39.9k'] = round((census_df['35k_to_39.999k']/census_df['total']*100),2)
census_df['pct_40k_to_44.9k'] = round((census_df['40k_to_44.999k']/census_df['total']*100),2)
census_df['pct_45k_to_49.9k'] = round((census_df['45k_to_49.999k']/census_df['total']*100),2)
census_df['pct_50k_to_59.9k'] = round((census_df['50k_to_59.999k']/census_df['total']*100),2)
census_df['pct_60k_to_74.9k'] = round((census_df['60k_to_74.999k']/census_df['total']*100),2)
census_df['pct_75k_to_99.9k'] = round((census_df['75k_to_99.999k']/census_df['total']*100),2)
census_df['pct_100k_to_124.9k'] = round((census_df['100k_to_124.999k']/census_df['total']*100),2)
census_df['pct_125k_to_149.9k'] = round((census_df['125k_to149.999k']/census_df['total']*100),2)
census_df['pct_150k_to_199.9k'] = round((census_df['150k_to_199.999k']/census_df['total']*100),2)
census_df['pct_200k_and_up'] = round((census_df['200k_or_more']/census_df['total']*100),2)


```
